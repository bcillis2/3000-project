# Preprocessing

## Drop single-valued data

```python
single_valued_cols = []
for i, col in enumerate(data.columns):
    if(data[col].nunique() == 1):
        single_valued_cols.append(data.columns[i])

single_valued_cols
```

```python
for col in single_valued_cols:
    data.drop(col, axis = 1, inplace=True)
```

Here we are dopping single-valued columns that are filled with only zeroes from the dataset. As we can see these are all weapons and they are single-valued because they are never used. Since they are never used we believe removing these columns is reasonable in this context.

## Encode Non-Numeric Data

```python
non_numeric_cols = [column for column in data.columns if data[column].dtype != 'float64']
#basically just grab any value that isn't a number (string, boolean, etc..)
non_numeric_cols
```

```python
# switch the bomb_planted and round_winner into 0 or 1 instead of true or false
le = LabelEncoder()
data["round_winner"] = le.fit_transform(data["round_winner"])
data["bomb_planted"] = le.fit_transform(data["bomb_planted"])

#One hot encode map column and keep map names
OHE = OneHotEncoder(handle_unknown="ignore", sparse=False)
transformed_data =OHE.fit_transform(data[['map']])
feature_names = OHE.get_feature_names_out(['map'])
OH_cols = pd.DataFrame(transformed_data, columns=feature_names, index=data.index)
data = data.join(OH_cols)
data.drop(['map'], axis=1, inplace=True)


print(data.head)
```

We encode non-numeric columns, we use label encoding and one-hot encoding. The three columns we encoded are "map", "bomb_planted", and "round_winner". We use label encoding on the bomb_planted and round_winner columns because these are binary values (True or False for bomb_planeted and T or CT for round_winner). One-hot encoding is used on the map column since there are multiple possible maps.

# Scaling

```python
target_col = "round_winner"
y_raw = data[target_col]
X_raw = data.drop(columns=[target_col])

scaler = StandardScaler()
X_scaled = pd.DataFrame(scaler.fit_transform(X_raw), columns=X_raw.columns) # want to keep the column names
```

Scaling the data for our logistic regression model. Kept raw values as well since decision tree and random forest classifier models do not require scaling.

# Feature Engineering

### Creating Features

```python
X_raw['ct_has_armor'] = [0 if row == 0 else 1 for row in X_raw['ct_armor']]
X_scaled['ct_has_armor'] = [0 if row == 0 else 1 for row in X_scaled['ct_armor']]
X_scaled.drop(columns=['ct_armor'])

X_raw['t_has_armor'] = [0 if row == 0 else 1 for row in X_raw['t_armor']]
X_scaled['t_has_armor'] = [0 if row == 0 else 1 for row in X_scaled['t_armor']]
X_scaled.drop(columns=['t_armor'])

display(X_raw.head())
display(X_scaled.head())
```

We create a team_has_armor feature for both teams to simplify for the model if one team has armor or not. We drop the original team_armor features for the scaled values because the scaled values will only be used by the Logistic Regression model and with Logistic Regression Models we might have to worry about multicollinearity (features being too similar) but Decision Trees handle this redundancy well.

# Train Test Split

```python
#this is for the decision tree and random forest classifier model
X_train_raw, X_test_raw, y_train_raw, y_test_raw = train_test_split(X_raw, y_raw, test_size=0.2, random_state=42)

#scaled data for the LR model
X_train_scaled, X_test_scaled, y_train, y_test = train_test_split(X_scaled, y_raw, test_size=0.2, random_state=42)
```

# Model Training and Evaluation

```python
from sklearn.metrics import (
    confusion_matrix, accuracy_score, precision_score, recall_score,
    f1_score, roc_curve, roc_auc_score
)

def evaluate_model(y_pred, y_test):
    print("ACCURACY: ", accuracy_score(y_test,y_pred))
    print("PRECISION: ", precision_score(y_test,y_pred))
    print("F1 SCORE: ", f1_score(y_test,y_pred))
    print("RECALL: ", recall_score(y_test,y_pred))

    
    CM = confusion_matrix(y_test,y_pred)
    CM_df = pd.DataFrame(CM, ['CT','T'], columns=['CT','T'])
    plt.figure(figsize=(12,8))
    sns.heatmap(CM_df, annot=True, fmt='d', cmap='Blues', cbar=False)
    plt.show()
```

### Logistic Regression Model

```python
LRM = LogisticRegression(random_state=42, C=2, max_iter=500)
LRM.fit(X_train_scaled, y_train)
LRM_y_pred = LRM.predict(X_test_scaled)
evaluate_model(LRM_y_pred, y_test)
```

### Decision Tree Model

```python
from sklearn.tree import DecisionTreeClassifier
DTM = DecisionTreeClassifier(random_state=42)
DTM.fit(X_train_raw,y_train_raw)
DTM_y_pred = DTM.predict(X_test_raw)
evaluate_model(DTM_y_pred, y_test_raw)
```

### Random Forest Classifier Model

```python
from sklearn.ensemble import RandomForestClassifier
RFM = RandomForestClassifier(random_state=42)
RFM.fit(X_train_raw, y_train_raw)
RFM_y_pred = RFM.predict(X_test_raw)
evaluate_model(RFM_y_pred, y_test_raw)
```

#### Random Forest Variable Feature Importance

```python
importances = RFM.feature_importances_
indices = np.argsort(importances)[::-1]

top_n = 20
indices = indices[:top_n]

f, ax = plt.subplots(figsize=(10, 8))
plt.title("Top 20 Variable Importance - Random Forest (CS:GO Data)")
sns.set_color_codes("pastel")

sns.barplot(
    y=[X_train_raw.columns[i] for i in indices], 
    x=importances[indices], 
    color="b"
)

ax.set(ylabel="Feature", xlabel="Variable Importance (Mean Decrease in Impurity)")
sns.despine(left=True, bottom=True)
plt.show()
```

### How does max depth affect accuracy?

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

depths = range(1, 31)

train_scores = []
test_scores = []

# Loop through each depth, train a model, and save scores
for d in depths:
    rf = RandomForestClassifier(max_depth=d, random_state=42, n_jobs=-1)
    rf.fit(X_train_raw, y_train_raw)
    
    # Evaluate on Train (to see if it's memorizing)
    train_pred = rf.predict(X_train_raw)
    train_scores.append(accuracy_score(y_train_raw, train_pred))
    
    # Evaluate on Test (to see if it generalizes)
    test_pred = rf.predict(X_test_raw)
    test_scores.append(accuracy_score(y_test_raw, test_pred))

# Plot the results
plt.figure(figsize=(10, 6))
plt.plot(depths, train_scores, label="Train Accuracy", marker='o')
plt.plot(depths, test_scores, label="Test Accuracy", marker='o', color='orange')

plt.xlabel("Max Depth of Trees")
plt.ylabel("Accuracy")
plt.title("Random Forest Complexity Curve (Depth vs Accuracy)")
plt.legend()
plt.grid(True)
plt.show()
```
